```
import matplotlib.pyplot as plt
import numpy as np
```

```
x1 = np.linspace(-3.0, 3.0, 100)
x2 = np.linspace(-3.0, 3.0, 100)
```

```
xx1, xx2 = np.meshgrid(x1,x2)
```

```
# Linear Function

fig, ax = plt.subplots(1,2)
fig.set_size_inches(15, 5)
fig.colorbar(ax[0].contourf(xx1,xx2, xx1+xx2 +10))
fig.colorbar(ax[1].contour(xx1,xx2, xx1+xx2 +10))
```

```
# Quadratic Function

fig, ax = plt.subplots(1,2)
fig.set_size_inches(15, 5)
fig.colorbar(ax[0].contourf(xx1,xx2, xx1**2+xx2**2 -10))
fig.colorbar(ax[1].contour(xx1,xx2, xx1**2+xx2**2 -10))
     
```

```
# Cubic Function

fig, ax = plt.subplots(1,2)
fig.set_size_inches(15, 5)
fig.colorbar(ax[0].contourf(xx1,xx2, xx1**3+xx2**3 -10))
fig.colorbar(ax[1].contour(xx1,xx2, xx1**3+xx2**3 -10))
     
```

```
# semi cubic

fig, ax = plt.subplots(1,2)
fig.set_size_inches(15, 5)
fig.colorbar(ax[0].contourf(xx1,xx2, xx1**3+xx2**2 -10))
fig.colorbar(ax[1].contour(xx1,xx2, xx1**3+xx2**2 -10))
     
```

```
# Fourth Power Function

fig, ax = plt.subplots(1,2)
fig.set_size_inches(15, 5)
fig.colorbar(ax[0].contourf(xx1,xx2, xx1**4-xx2**2 -10))
fig.colorbar(ax[1].contour(xx1,xx2, xx1**4-xx2**2 -10))
     
```

```
 # Higher order polynomial

fig, ax = plt.subplots(1,2)
fig.set_size_inches(15, 5)
fig.colorbar(ax[0].contourf(xx1,xx2, xx1**8 + xx1**7 + xx1**6 + xx1**4-xx2**2 -10))
fig.colorbar(ax[1].contour(xx1,xx2, xx1**8 + xx1**7 + xx1**6 + xx1**4-xx2**2 -10))
     
```
